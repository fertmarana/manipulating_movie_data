```python
#separando em chunks porque é muito grande pra rodar a funcao pivot_table sozinha
chunk_size = 50000
chunks = [x for x in range(0, ratings.shape[0], chunk_size)]
print(ratings.shape[0])
print(chunks)
#for i in range(0, len(chunks) - 1):
    #print(chunks[i], chunks[i + 1] - 1)
```

```python
# agora que separamos em chunks vamos ir adicionando pouco a pouco na tabela pivot 
from tqdm import tqdm

pivot_df = pd.DataFrame()
for i in tqdm(range(0, len(chunks) - 1)):
  chunk_df = ratings.iloc[ chunks[i]:chunks[i + 1] - 1]
  interactions = (chunk_df.groupby(["user_id", "movie_id"])["rating"]
      .sum()
      .unstack()
      .reset_index()
      .fillna(0)
      .set_index("user_id")
    )
  print (interactions.shape)
  pivot_df = pivot_df.append(interactions, sort=False) 
```
